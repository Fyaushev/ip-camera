```python
from selenium import webdriver
import time

options = webdriver.ChromeOptions()
options.add_argument('headless')
options.add_argument("--no-sandbox")
options.add_argument('window-size=1920x1080')
options.add_argument("disable-gpu")

# Start Chrome Driver
chromedriver = 'utils/chromedriver'

driver = webdriver.Chrome(chromedriver, options=options)

# Open the URL you want to execute JS
URL = 'http://166.145.68.221/view/view.shtml'
driver.get(URL)
```

```python
# Execute JS

driver.execute_script("continousMove('tilt', '0,-100')")
time.sleep(1)
driver.execute_script("continousMove('tilt', '0,0')")
```

```python
driver.execute_script("continousMove('zoom', '-200')")
time.sleep(1)
driver.execute_script("continousMove('zoom', '0')")
```

```python
driver.quit()
```
